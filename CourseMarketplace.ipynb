```python
import requests
from bs4 import BeautifulSoup
from csv import reader
import csv
from urllib.parse import urlparse

with open('source2.csv', 'r') as read_obj:
    
    csv_reader = reader(read_obj)
       
    dyn = list(csv_reader);
    
    with open('resultsudemy.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Source", "Link", "Title", "Topic", "Star", "Review","Student", "Creator"])
        
        i = 1
        for elem in dyn:
            #for elem in row:
                
                alamat = elem[0]
                res = requests.get(alamat)     # menggunakan library request dengan mengirimkan header
                html = BeautifulSoup(res.content, "html.parser")    # parsing html menggunakan Beautifulsoup
                
                url = urlparse(alamat).netloc
                domain = '.'.join(url.split('.')[1:]) # mengambil nama domain
                
                try:
                    
                    if (domain == "udemy.com"):
                        title = html.find("h1", class_  = "udlite-heading-xl clp-lead__title clp-lead__title--small").get_text().replace('\n', '')
                        star = html.find("span", class_ = "udlite-heading-sm star-rating--rating-number--2o8YM").get_text().replace('\n', '')
                        topic = html.find("a", class_   = "udlite-heading-sm").get_text().replace('\n', '')
                        topic2 = html.find_all("a",{"class":"udlite-heading-sm"})[1]
                        topic3 = topic2.get_text().replace('\n', '')
                        totalrating = html.find_all("a",{"class":"styles--rating-wrapper--5a0Tr"})
                        totalrating2 = totalrating[0].find_all("span")[3].get_text().replace('\n', '')
                        totalrating3 = totalrating2.replace("rating)","")
                        totalstudent = html.find_all(attrs={"data-purpose": "enrollment"})
                        totalstudent2 = totalstudent[0].get_text().replace('\n', '')
                        totalstudent3 = totalstudent2.replace("peserta","")
                        creator = html.find("a", class_ = "udlite-btn udlite-btn-large udlite-btn-link udlite-heading-md udlite-text-sm udlite-instructor-links").get_text().replace('\n', '')


                    else:
                        title = "Link Tidak Ditemukan"
                        star = "Link Tidak Ditemukan"

                    if (title != "Link Tidak Ditemukan"):
                        writer.writerow([domain, alamat, title, topic3, star, totalrating3, totalstudent3, creator])

                    print("(",i,")", domain,"-", title,"-",star, topic3, totalstudent3, creator,totalrating3)
                    i += 1

                except:
                    #writer.writerow(["EROR", "EROR", "EROR", "EROR"])
                    print("(",i,")", "eror!")
                    i += 1
                    pass
    
```
